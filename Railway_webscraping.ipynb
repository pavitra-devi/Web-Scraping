```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from lxml import html
```

```python
#url="https://www.makemytrip.com/railways/vijayawada-jn-bza-railway-station.html"
#response=requests.get(url)
#print(response)
```

```python
f=open("railways.html",'r')
response=f.read()
print(response)
```

```python
soup=BeautifulSoup(response,"html.parser")
print(soup)
```

```python
title=soup.find("a",{"class":"seoName"})
title.string
```

```python
all_titles=soup.find_all('a',{"class":"seoName"})
print(all_titles)
```

```python
titles=[]
for i in all_titles:
  name=i.string
  titles.append(name)
print(titles)
```

```python
print(len(titles))
```

```python
all_train_num=soup.find_all('span',{"class":"railSeoWrap__infoTable--railNo"})
print(all_train_num)
print(len(all_train_num))
```

```python
train_num=[]
for i in all_train_num:
  name=i.get_text()
  train_num.append(name[2:])
print(train_num)
```

```python
print(len(train_num))
```

```python
Running_day=soup.find_all('span',{'class':"latoBold weeklySchedule"})
print(Running_day)
```

```python
print(len(Running_day))
```

```python
Running_days=[]
for i in Running_day:
  days=""
  day=i.find_all('span',{'class':"greenText"})
  for j in day:
    #print(j,end=" ")
    name=j.attrs
    name=name['title']+","
    #print(name,end=" ")
    days+=name
  Running_days.append(days)
print(Running_days)
```

```python
print(len(Running_days))
```

```python
source_departures=soup.find_all('span',{'class':"railSeoWrap__infoTable--startTime"})
print(source_departures)
```

```python
source_departure=[]
for i in source_departures:
  name=i.get_text()
  #if name!=None:
  source_departure.append(name)
print(source_departure)
```

```python
destination_arrivals=soup.find_all('span',{'class':"railSeoWrap__infoTable--endTime"})
print(destination_arrivals)
```

```python
destination_arrival=[]
for i in destination_arrivals:
  name=i.get_text()
  #if name!=None:
  destination_arrival.append(name)
print(destination_arrival)
```

```python
print(len(source_departure),len(destination_arrival))
```

```python
sources=soup.find_all('span',{"class":"railSeoWrap__infoTable--stnFullNm"})
print(sources)
```

```python
source=[]
destination=[]
j=0
for i in sources:
  name=i.get_text()
  if j%2==0:
    source.append(name)
  else:
    destination.append(name)
  j+=1
print(source)
```

```python
print(destination)
```

```python
print(len(source),len(destination))
```

```python
codes=soup.find_all('span',{"class":"railSeoWrap__infoTable--stnShrtName"})
print(codes)
```

```python
source_code=[]
destination_code=[]
j=0
for i in codes:
  name=i.get_text()
  if j%2==0:
    source_code.append(name)
  else:
    destination_code.append(name)
  j+=1
print(source_code)
```

```python
print(destination_code)
```

```python
print(len(source_code),len(destination_code))
```

```python
durations=soup.find_all('span',{"class":"appendLeft5 appendRight5 duration"})
print(durations)
```

```python
duration=[]
for i in durations:
  name=i.get_text()
  #if name!=None:
  duration.append(name)
print(duration)
```

```python
print(len(duration))
```

```python
stops_distances=soup.find_all(attrs={"class":"stp_type darkGreyText"})
print(stops_distances)
```

```python
print(len(stops_distances))
```

```python
stops_distance=[]
for i in stops_distances:
  name=i.get_text()
  #if name!=None:
  stops_distance.append(name)
print(stops_distance)
```

```python
print(len(stops_distance))
```
